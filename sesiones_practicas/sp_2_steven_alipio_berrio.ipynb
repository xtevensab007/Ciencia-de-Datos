### <span style="color:#2F749F;"><strong>Ejercicio 3: C치lculo de medidas de distribuci칩n</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza la distribuci칩n de los datos. Calcula:
1. Asimetr칤a
2. Curtosis
3. Realiza la comparaci칩n con la distribuci칩n normal (ver la imagen de ejemplo)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# Definir la semilla y par치metros
base = 661504  # 칰ltimos 6 d칤gitos
np.random.seed(base)

media_teorica = (base % 10) + 2
desviacion_estandar = 2
tamano_muestra = 30

# Simular los ingresos
ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)
ingresos = np.where(ingresos < 0, 0, ingresos)  # valores negativos a 0

# Calcular medidas de tendencia central
media = np.mean(ingresos)
mediana = np.median(ingresos)
moda_result = stats.mode(ingresos, keepdims=True)
moda = round(moda_result.mode[0])

# Calcular asimetr칤a y curtosis
asimetria = stats.skew(ingresos)
curtosis = stats.kurtosis(ingresos)

# Graficar distribuci칩n con comparaci칩n normal
plt.figure(figsize=(12, 6))
sns.histplot(ingresos, bins=10, kde=True, color="skyblue", edgecolor="black", alpha=0.4)

# L칤nea de la media
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')

# Distribuci칩n normal te칩rica ajustada a los datos
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, np.std(ingresos))
plt.plot(x, p * tamano_muestra * (xmax - xmin) / 10, 'b-', lw=2)

# Mostrar asimetr칤a y curtosis en el gr치fico
plt.text(media + 0.5, max(np.histogram(ingresos, bins=10)[0]) * 0.8,
         f"Asimetr칤a: {asimetria:.2f}\nCurtosis: {curtosis:.2f}",
         bbox=dict(facecolor='white', alpha=0.7))

# Configuraci칩n de gr치fico
plt.title("Distribuci칩n de la muestra vs. distribuci칩n normal")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# Imprimir resultados
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Asimetr칤a: {asimetria:.2f}")
print(f"Curtosis: {curtosis:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# Reutilizamos la muestra del Ejercicio 1

base = 661504
np.random.seed(base)

media_teorica = (base % 10) + 2
desviacion_estandar = 2
tamano_muestra = 1000  # aumentamos tama침o para ver mejor la curva
ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)
ingresos = np.where(ingresos < 0, 0, ingresos)


# Calcular Percentiles, Deciles y Cuartiles

percentiles = np.percentile(ingresos, [10, 25, 50, 75, 90])  # P10, P25, P50, P75, P90
deciles = np.percentile(ingresos, np.arange(10, 100, 10))   # D1 a D9
cuartiles = np.percentile(ingresos, [25, 50, 75])           # Q1, Q2, Q3


# Graficar distribuci칩n con l칤neas

plt.figure(figsize=(14, 6))
sns.histplot(ingresos, bins=20, kde=True, color="skyblue", edgecolor="black", alpha=0.3)

# Dibujar percentiles principales
for p, val in zip([10, 25, 50, 75, 90], percentiles):
    plt.axvline(val, color='blue', linestyle='-', linewidth=1.5)
    plt.text(val, plt.ylim()[1]*0.9, f'P{p}\n', rotation=90, color='blue', va='top', ha='center')

# Dibujar deciles
for i, val in enumerate(deciles, 1):
    plt.axvline(val, color='gray', linestyle='--', linewidth=1)
    plt.text(val, plt.ylim()[1]*0.6, f'D{i}', rotation=90, color='gray', va='top', ha='center')

# T칤tulo y etiquetas
plt.title("Percentiles, Deciles y Cuartiles sobre la Distribuci칩n")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")

# Cuadro explicativo
plt.legend([f"P10={percentiles[0]:.2f}, P25={percentiles[1]:.2f}, "
            f"P50={percentiles[2]:.2f}, P75={percentiles[3]:.2f}, "
            f"P90={percentiles[4]:.2f}"], loc="upper right")

# Cuadro de texto explicativo
plt.text(np.min(ingresos), plt.ylim()[1]*0.95,
         "Percentiles: posici칩n relativa (P10, P25, P50, P75, P90)\n"
         "Deciles: dividen en 10 partes (D1 a D9)\n"
         "Cuartiles: Q1 = 25%, Q2 = 50%, Q3 = 75%\n"
         "Analizar dispersi칩n, concentraci칩n y desigualdad",
         bbox=dict(facecolor='white', alpha=0.7), ha="left", va="top")

plt.show()


# Imprimir resultados

print("Percentiles (P10, P25, P50, P75, P90):", percentiles)
print("Deciles (D1 a D9):", deciles)
print("Cuartiles (Q1, Q2, Q3):", cuartiles)
```

## <span style="color:black;"><strong>Parte 2: An치lisis Bivariado</strong></span>

### <span style="color:#2F749F;"><strong>Ejercicio 5: An치lisis de correlaci칩n</strong></span>

En esta segunda parte, genera **dos variables** a partir de t칰 n칰mero de c칠dula y realiza un an치lisis **bivariado** para calcular la relaci칩n entre ellas.

1. Use los **6 칰ltimos d칤gitos** de su c칠dula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribuci칩n normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlaci칩n de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

<p align="center">
    <img src="imgs\sp_2_e5.png" alt="sp_2_e5" width="900"  height="400">
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns


# Definir la semilla

base = 661504
np.random.seed(base)


# Generar dos variables aleatorias

n = 100  # tama침o de la muestra

# Ejemplo: ingresos y gastos
ingresos = np.random.normal(loc=(base % 50) + 20, scale=10, size=n)  # media dependiente de la c칠dula
gastos = ingresos * 0.6 + np.random.normal(loc=10, scale=5, size=n)  # correlaci칩n positiva con ruido


# Calcular coeficientes de correlaci칩n

pearson_corr, p_pearson = stats.pearsonr(ingresos, gastos)
spearman_corr, p_spearman = stats.spearmanr(ingresos, gastos)


# Gr치fico de dispersi칩n con l칤nea de regresi칩n

plt.figure(figsize=(10,6))
sns.regplot(x=ingresos, y=gastos, ci=None, scatter_kws={"alpha":0.6, "color":"skyblue"}, line_kws={"color":"red"})

# Etiquetas y t칤tulo
plt.title("Relaci칩n entre Ingresos y Gastos")
plt.xlabel("Ingresos")
plt.ylabel("Gastos")

# Texto con resultados
plt.text(np.min(ingresos), np.max(gastos),
         f"Coef. Pearson: {pearson_corr:.2f} (p={p_pearson:.3f})\n"
         f"Coef. Spearman: {spearman_corr:.2f} (p={p_spearman:.3f})",
         bbox=dict(facecolor='white', alpha=0.7))

plt.grid(True)
plt.tight_layout()
plt.show()


# Imprimir resultados

print(f"Coeficiente de correlaci칩n de Pearson: {pearson_corr:.2f}, p-valor: {p_pearson:.3f}")
print(f"Coeficiente de correlaci칩n de Spearman: {spearman_corr:.2f}, p-valor: {p_spearman:.3f}")
```

Coeficientes de correlaci칩n:

Pearson = 0.66 (p = 0.000): indica una correlaci칩n positiva moderada-alta y significativa.
游녤 A mayor ingreso, tienden a aumentar los gastos.

Spearman = 0.67 (p = 0.000): tambi칠n positivo y significativo.
游녤 Confirma que la relaci칩n es consistente incluso si no es lineal perfecta.

Como ambos son muy parecidos, esto refuerza que la relaci칩n entre ingresos y gastos es clara y robusta.

Dispersi칩n de los datos:

La nube de puntos muestra una tendencia ascendente: quienes tienen mayores ingresos, en general, gastan m치s.

Sin embargo, no todos los puntos caen exactamente sobre la l칤nea de tendencia  hay variabilidad individual (no todos gastan lo mismo al ganar lo mismo).

L칤nea de regresi칩n (en rojo):

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociaci칩n estad칤stica** entre dos variables categ칩ricas simuladas a partir de tu n칰mero de c칠dula.

1. Usa los **6 칰ltimos d칤gitos** de tu c칠dula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categ칩ricas:
   - `Nivel educativo`: B치sico, Medio, Superior.
   - `Acceso a internet`: S칤, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribuci칩n conjunta con un mapa de calor (ver la imagen de con la informaci칩n de ejemplo).
6. Interpreta el resultado.

    <img src="imgs\sp_2_e6.png" alt="sp_2_e6" width="900"  height="700">

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency


# Definir la semilla

base = 661504
np.random.seed(base)


# Simular datos categ칩ricos

n = 60  # tama침o de la muestra

# Variable 1: Nivel educativo
niveles = ["B치sico", "Medio", "Superior"]
nivel_educativo = np.random.choice(niveles, size=n, p=[0.3, 0.4, 0.3])

# Variable 2: Acceso a internet
opciones_internet = ["S칤", "No"]
acceso_internet = np.random.choice(opciones_internet, size=n, p=[0.7, 0.3])

# Crear DataFrame
data = pd.DataFrame({
    "Nivel Educativo": nivel_educativo,
    "Acceso a Internet": acceso_internet
})


# Tabla de contingencia

tabla = pd.crosstab(data["Nivel Educativo"], data["Acceso a Internet"])


# Prueba Chi-cuadrado

chi2, p, dof, expected = chi2_contingency(tabla)


# Visualizaci칩n con mapa de calor

plt.figure(figsize=(8,6))
sns.heatmap(tabla, annot=True, fmt="d", cmap="Blues")
plt.title("Distribuci칩n conjunta: Nivel educativo vs. Acceso a Internet")
plt.ylabel("Nivel educativo")
plt.xlabel("Acceso a Internet")
plt.show()


# Resultados

print("Tabla de contingencia:\n", tabla, "\n")
print(f"Chi-cuadrado: {chi2:.2f}")
print(f"Grados de libertad: {dof}")
print(f"P-valor: {p:.4f}")
print("Frecuencias esperadas:\n", expected)
```

Distribuci칩n observada:

Nivel B치sico: 4 sin internet, 14 con internet.

Nivel Medio: 3 sin internet, 18 con internet.

Nivel Superior: 5 sin internet, 16 con internet.

Se ve claramente que, en todos los niveles educativos, la mayor칤a tiene acceso a internet.

Interpretaci칩n del mapa de calor:

Los cuadros m치s oscuros est치n en la columna S칤 (acceso a internet), lo que muestra que la mayor parte de los estudiantes/personas accede a internet.

El nivel Medio con internet (18) es el grupo m치s numeroso.

Los casos sin internet son pocos en comparaci칩n, independientemente del nivel educativo.

Prueba Chi-cuadrado (independencia):

Con esta tabla se calcula el estad칤stico 픦 para verificar si el acceso a internet depende del nivel educativo.

Generalmente, si el p-valor > 0.05, se concluye que no hay relaci칩n significativa  el acceso a internet no est치 condicionado por el nivel educativo.

Si el p-valor 곣 0.05, se concluye que s칤 existe asociaci칩n  el acceso a internet var칤a seg칰n el nivel educativo.

La mayor칤a de las personas, sin importar su nivel educativo, tiene acceso a internet.

Aunque el acceso parece un poco m치s alto en niveles Medio y Superior, la diferencia no es muy marcada.

El test Chi-cuadrado permitir칤a confirmar si esta diferencia es estad칤sticamente significativa o no.

En la pr치ctica, esto refleja que hoy en d칤a el internet es bastante com칰n en todos los niveles educativos, aunque a칰n existen peque침os grupos sin acceso.
