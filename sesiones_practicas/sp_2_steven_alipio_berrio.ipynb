### <span style="color:#2F749F;"><strong>Ejercicio 1: Cálculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el número base `528310` (debes reemplazarlo con los últimos 6 dígitos de tu documento de identificación).  
2. Con este número genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribución normal:  
   - Media teórica: `(base % 10) + 2`  
   - Desviación estándar: `2`  
   - Tamaño de muestra: `30`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# Definir la semilla y parámetros
base = 661504 #últimos 6 dígitos
np.random.seed(base)

media_teorica = (base % 10) + 2
desviacion_estandar = 2
tamano_muestra = 30

#Simular los ingresos
ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)
ingresos = np.where(ingresos < 0, 0, ingresos)  # valores negativos a 0

#Calcular media, mediana y moda
media = np.mean(ingresos)
mediana = np.median(ingresos)
moda_result = stats.mode(ingresos, keepdims=True)
moda = round(moda_result.mode[0])
#moda = np.round(moda_result.mode[0], 0)
#moda = moda_result.mode[0]

#Graficar
plt.figure(figsize=(10, 6))
plt.hist(ingresos, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
sns.histplot(ingresos, bins=10, kde=True, color="skyblue", edgecolor="black", alpha=0.4)
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='orange', linestyle='-.', label=f'Moda: {moda:.2f}')
plt.title("Distribución de Ingresos Simulados")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Imprimir los valores
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
```

Media (6.30)
La media representa el valor promedio de los ingresos de la muestra. Nos dice que, en promedio, los ingresos rondan los 6.3 unidades.

Mediana (6.31)
La mediana es el valor central cuando ordenamos los datos. En este caso, es casi igual a la media (6.31), lo que indica que la distribución está bastante simétrica y no hay valores extremos que desvíen demasiado el promedio.

Moda (2.00)
La moda es el valor más frecuente. Aquí aparece en 2.0, lo que significa que hubo varios individuos con ese ingreso, aunque no sea cercano al promedio general. Esto muestra que en la muestra hay un pequeño grupo con ingresos bajos que se repiten varias veces.

Forma de la distribución

El histograma muestra una forma relativamente simétrica, pero con una ligera concentración hacia la derecha (valores mayores a 6).

No se observan valores muy extremos (outliers), pero sí cierta dispersión en los ingresos.

Los datos se distribuyen de forma bastante equilibrada.

Como la media ≈ mediana, podemos decir que la muestra no presenta sesgo fuerte.

La moda más baja sugiere que hay un grupo de personas con ingresos más reducidos que se repiten, aunque la mayoría se concentra alrededor de los 6–8.

### <span style="color:#2F749F;"><strong>Ejercicio 2: Cálculo de medidas de dispersión</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza qué tan dispersos están los datos respecto a su media. Calcula:
1. Rango
2. Varianza
3. Desviación estándar
4. Coeficiente de variación
5. Genera una gráfica con los resultados obtenidos (ver imagen de ejemplo)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Generar muestra (como en Ejercicio 1)
base = 661504
np.random.seed(base)

media_teorica = (base % 10) + 2
ingresos = np.random.normal(loc=media_teorica, scale=2, size=30)
ingresos = np.where(ingresos < 0, 0, ingresos)


# Calcular medidas de dispersión
media = np.mean(ingresos)
minimo = np.min(ingresos)
maximo = np.max(ingresos)
rango = maximo - minimo
varianza = np.var(ingresos, ddof=1)
desviacion = np.std(ingresos, ddof=1)
coef_var = desviacion / media


# Crear la gráfica
plt.figure(figsize=(12, 6))
sns.histplot(ingresos, bins=10, kde=True, color="skyblue", edgecolor="black", alpha=0.4)

# Líneas de media y ±1 desviación estándar
plt.axvline(media, color='blue', linestyle='-', linewidth=2, label=f'Media = {media:.2f}')
plt.axvline(media + desviacion, color='purple', linestyle='--', label=f'+1σ = {media + desviacion:.2f}')
plt.axvline(media - desviacion, color='purple', linestyle='--', label=f'-1σ = {media - desviacion:.2f}')
plt.axvline(minimo, color='red', linestyle='--', label=f'Mínimo = {minimo:.2f}')
plt.axvline(maximo, color='red', linestyle='--', label=f'Máximo = {maximo:.2f}')

# Recuadro con las estadísticas
texto = (f"Rango: {rango:.2f}\n"
         f"Varianza: {varianza:.2f}\n"
         f"Desviación estándar: {desviacion:.2f}\n"
         f"Coef. de variación: {coef_var:.2f}")
plt.text(0.02, 0.95, texto, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='lightgreen', edgecolor='black'))

# Etiquetas y detalles
plt.title("Distribución de ingresos simulados — Medidas de dispersión")
plt.xlabel("Ingreso mensual (miles de pesos)")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```

Rango (7.15)

El rango es la diferencia entre el valor máximo y el mínimo (9.02 – 1.87). Nos indica el intervalo total de variación de los ingresos. Aquí vemos que los ingresos más bajos están alrededor de 1.9 y los más altos en 9.0, por lo que hay bastante amplitud en la muestra.

La desviación estándar, más fácil de interpretar, nos dice que en promedio los ingresos se alejan 1.75 unidades del valor medio (6.30).
Esto significa que los ingresos no son todos iguales, pero tampoco presentan variaciones extremas.

Coeficiente de variación (0.28)
Este valor es la razón entre la desviación estándar y la media.

Un valor bajo (<0.30) indica baja dispersión relativa.

En este caso (0.28), los ingresos son relativamente homogéneos respecto a su promedio.

Valores mínimos y máximos

Mínimo: 1.87

Máximo: 9.02
Reflejan los extremos de la muestra. Aunque hay ingresos bajos y altos, la mayoría se concentra alrededor de la media.

Desviaciones estándar (±1σ)

Entre 4.55 y 8.04 está el rango que cubre una desviación estándar alrededor de la media.

### <span style="color:#2F749F;"><strong>Ejercicio 3: Cálculo de medidas de distribución</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza la distribución de los datos. Calcula:
1. Asimetría
2. Curtosis
3. Realiza la comparación con la distribución normal (ver la imagen de ejemplo)

image.png<p align="center">
    <img src="imgs\sp_2_e3.png" alt="sp_2_e3" width="900"  height="400">
</p>



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# Definir la semilla y parámetros
base = 661504  # últimos 6 dígitos
np.random.seed(base)

media_teorica = (base % 10) + 2
desviacion_estandar = 2
tamano_muestra = 30

# Simular los ingresos
ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)
ingresos = np.where(ingresos < 0, 0, ingresos)  # valores negativos a 0

# Calcular medidas de tendencia central
media = np.mean(ingresos)
mediana = np.median(ingresos)
moda_result = stats.mode(ingresos, keepdims=True)
moda = round(moda_result.mode[0])

# Calcular asimetría y curtosis
asimetria = stats.skew(ingresos)
curtosis = stats.kurtosis(ingresos)

# Graficar distribución con comparación normal
plt.figure(figsize=(12, 6))
sns.histplot(ingresos, bins=10, kde=True, color="skyblue", edgecolor="black", alpha=0.4)

# Línea de la media
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')

# Distribución normal teórica ajustada a los datos
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, np.std(ingresos))
plt.plot(x, p * tamano_muestra * (xmax - xmin) / 10, 'b-', lw=2)

# Mostrar asimetría y curtosis en el gráfico
plt.text(media + 0.5, max(np.histogram(ingresos, bins=10)[0]) * 0.8,
         f"Asimetría: {asimetria:.2f}\nCurtosis: {curtosis:.2f}",
         bbox=dict(facecolor='white', alpha=0.7))

# Configuración de gráfico
plt.title("Distribución de la muestra vs. distribución normal")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# Imprimir resultados
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Asimetría: {asimetria:.2f}")
print(f"Curtosis: {curtosis:.2f}")
```

Media (6.30)
La línea roja punteada marca la media de los ingresos simulados. La mayor parte de los datos se concentra cerca de este valor, lo que ya vimos en los análisis anteriores.

Asimetría (-0.66)
El valor negativo indica una asimetría hacia la izquierda, es decir, que hay algunos ingresos bajos que “jalan” un poco la distribución en esa dirección.
Sin embargo, el valor no es muy extremo (entre -1 y 1 se considera asimetría moderada), por lo que la distribución sigue siendo bastante cercana a simétrica.

Curtosis (-0.06)
La curtosis mide qué tan “aplanada” o “picuda” es la distribución comparada con la normal.

Un valor cercano a 0 indica que la forma es muy parecida a la normal.

Aquí, con -0.06, prácticamente no hay diferencia: la distribución es casi mesocúrtica (similar a la campana de Gauss).

Comparación con la distribución normal (línea azul)
Al superponer la curva normal teórica sobre el histograma, se observa que los datos se ajustan bastante bien. Hay ligeras diferencias en los extremos (colas), pero en general la forma es consistente con una distribución normal.

La distribución de ingresos simulados se ajusta bastante bien a una distribución normal:

Es casi simétrica (ligera inclinación hacia la izquierda).

Tiene una curtosis muy cercana a la normal, lo que significa que no hay colas muy pesadas ni muy planas.

Esto confirma que los datos son adecuados para análisis estadísticos que asumen normalidad (como correlaciones o pruebas paramétricas).

### <span style="color:#2F749F;"><strong>Ejercicio 4: Cálculo de medidas de percentiles</strong></span>

A partir de la muestra generada en el Ejercicio 1, calcula los percentiles de tus datos. Calcula:
1. Percentiles
2. Deciles
3. Cuartiles
4. Grafica los anteriores percentiles calculados, sobre la distribución de tus datos (ver la imagen de ejemplo)

image.png<p align="center">
    


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# Reutilizamos la muestra del Ejercicio 1

base = 661504
np.random.seed(base)

media_teorica = (base % 10) + 2
desviacion_estandar = 2
tamano_muestra = 1000  # aumentamos tamaño para ver mejor la curva
ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)
ingresos = np.where(ingresos < 0, 0, ingresos)


# Calcular Percentiles, Deciles y Cuartiles

percentiles = np.percentile(ingresos, [10, 25, 50, 75, 90])  # P10, P25, P50, P75, P90
deciles = np.percentile(ingresos, np.arange(10, 100, 10))   # D1 a D9
cuartiles = np.percentile(ingresos, [25, 50, 75])           # Q1, Q2, Q3


# Graficar distribución con líneas

plt.figure(figsize=(14, 6))
sns.histplot(ingresos, bins=20, kde=True, color="skyblue", edgecolor="black", alpha=0.3)

# Dibujar percentiles principales
for p, val in zip([10, 25, 50, 75, 90], percentiles):
    plt.axvline(val, color='blue', linestyle='-', linewidth=1.5)
    plt.text(val, plt.ylim()[1]*0.9, f'P{p}\n', rotation=90, color='blue', va='top', ha='center')

# Dibujar deciles
for i, val in enumerate(deciles, 1):
    plt.axvline(val, color='gray', linestyle='--', linewidth=1)
    plt.text(val, plt.ylim()[1]*0.6, f'D{i}', rotation=90, color='gray', va='top', ha='center')

# Título y etiquetas
plt.title("Percentiles, Deciles y Cuartiles sobre la Distribución")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")

# Cuadro explicativo
plt.legend([f"P10={percentiles[0]:.2f}, P25={percentiles[1]:.2f}, "
            f"P50={percentiles[2]:.2f}, P75={percentiles[3]:.2f}, "
            f"P90={percentiles[4]:.2f}"], loc="upper right")

# Cuadro de texto explicativo
plt.text(np.min(ingresos), plt.ylim()[1]*0.95,
         "Percentiles: posición relativa (P10, P25, P50, P75, P90)\n"
         "Deciles: dividen en 10 partes (D1 a D9)\n"
         "Cuartiles: Q1 = 25%, Q2 = 50%, Q3 = 75%\n"
         "Analizar dispersión, concentración y desigualdad",
         bbox=dict(facecolor='white', alpha=0.7), ha="left", va="top")

plt.show()


# Imprimir resultados

print("Percentiles (P10, P25, P50, P75, P90):", percentiles)
print("Deciles (D1 a D9):", deciles)
print("Cuartiles (Q1, Q2, Q3):", cuartiles)
```

Percentiles

P10 = 3.33 → El 10% de los ingresos está por debajo de 3.33.

P25 = 4.58 → El 25% de los ingresos no supera ese valor (primer cuartil, Q1).

P50 = 5.89 → El 50% de los ingresos se encuentra debajo de este valor (mediana, Q2).

P75 = 7.20 → El 75% de los ingresos está por debajo de este nivel (tercer cuartil, Q3).

P90 = 8.56 → El 90% de los ingresos está debajo de este valor, y solo el 10% lo supera.

Esto muestra cómo se distribuyen los ingresos a lo largo de la población.

Deciles (D1 a D9)
Los deciles dividen la distribución en 10 partes iguales. Nos ayudan a ver con más detalle cómo se concentran los ingresos.
Ejemplo: D5 coincide con la mediana, mientras que D9 está cerca de los valores más altos (8.56).

Cuartiles (Q1, Q2, Q3)

Q1 (25%): 4.58

Q2 (50%): 5.89

Q3 (75%): 7.20
Estos valores dividen la muestra en 4 partes iguales. La mayoría de los ingresos se concentran entre Q1 y Q3 (de 4.58 a 7.20).

Interpretación gráfica
La curva suavizada muestra que la distribución es bastante “normal”, con un pico en el centro.
La dispersión entre Q1 y Q3 refleja la zona donde está el 50% central de los datos.
Los percentiles más bajos y más altos muestran que existen ingresos pequeños y grandes, pero sin valores extremos exagerados.

El análisis con percentiles, deciles y cuartiles nos confirma que:

## <span style="color:black;"><strong>Parte 2: Análisis Bivariado</strong></span>

### <span style="color:#2F749F;"><strong>Ejercicio 5: Análisis de correlación</strong></span>

En esta segunda parte, genera **dos variables** a partir de tú número de cédula y realiza un análisis **bivariado** para calcular la relación entre ellas.

1. Use los **6 últimos dígitos** de su cédula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlación de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

<p align="center">
    <img src="imgs\sp_2_e5.png" alt="sp_2_e5" width="900"  height="400">
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns


# Definir la semilla

base = 661504
np.random.seed(base)


# Generar dos variables aleatorias

n = 100  # tamaño de la muestra

# Ejemplo: ingresos y gastos
ingresos = np.random.normal(loc=(base % 50) + 20, scale=10, size=n)  # media dependiente de la cédula
gastos = ingresos * 0.6 + np.random.normal(loc=10, scale=5, size=n)  # correlación positiva con ruido


# Calcular coeficientes de correlación

pearson_corr, p_pearson = stats.pearsonr(ingresos, gastos)
spearman_corr, p_spearman = stats.spearmanr(ingresos, gastos)


# Gráfico de dispersión con línea de regresión

plt.figure(figsize=(10,6))
sns.regplot(x=ingresos, y=gastos, ci=None, scatter_kws={"alpha":0.6, "color":"skyblue"}, line_kws={"color":"red"})

# Etiquetas y título
plt.title("Relación entre Ingresos y Gastos")
plt.xlabel("Ingresos")
plt.ylabel("Gastos")

# Texto con resultados
plt.text(np.min(ingresos), np.max(gastos),
         f"Coef. Pearson: {pearson_corr:.2f} (p={p_pearson:.3f})\n"
         f"Coef. Spearman: {spearman_corr:.2f} (p={p_spearman:.3f})",
         bbox=dict(facecolor='white', alpha=0.7))

plt.grid(True)
plt.tight_layout()
plt.show()


# Imprimir resultados

print(f"Coeficiente de correlación de Pearson: {pearson_corr:.2f}, p-valor: {p_pearson:.3f}")
print(f"Coeficiente de correlación de Spearman: {spearman_corr:.2f}, p-valor: {p_spearman:.3f}")
```

Coeficientes de correlación:

Pearson = 0.66 (p = 0.000): indica una correlación positiva moderada-alta y significativa.
👉 A mayor ingreso, tienden a aumentar los gastos.

Spearman = 0.67 (p = 0.000): también positivo y significativo.
👉 Confirma que la relación es consistente incluso si no es lineal perfecta.

Como ambos son muy parecidos, esto refuerza que la relación entre ingresos y gastos es clara y robusta.

Dispersión de los datos:

La nube de puntos muestra una tendencia ascendente: quienes tienen mayores ingresos, en general, gastan más.

Sin embargo, no todos los puntos caen exactamente sobre la línea de tendencia → hay variabilidad individual (no todos gastan lo mismo al ganar lo mismo).

Línea de regresión (en rojo):

Marca la tendencia promedio: cada vez que aumentan los ingresos, los gastos también suben en proporción.

Esto sugiere que los gastos están fuertemente ligados al nivel de ingresos.

Existe una relación positiva y significativa entre ingresos y gastos.

Mientras los ingresos crecen, también lo hacen los gastos, lo cual es lógico y esperado.

La correlación no es perfecta, lo que refleja que algunas personas pueden gastar menos o más de lo que corresponde a sus ingresos.

En términos prácticos, este análisis demuestra que los ingresos influyen bastante en el nivel de gastos, pero no lo explican al 100%

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency


# Definir la semilla

base = 661504
np.random.seed(base)


# Simular datos categóricos

n = 60  # tamaño de la muestra

# Variable 1: Nivel educativo
niveles = ["Básico", "Medio", "Superior"]
nivel_educativo = np.random.choice(niveles, size=n, p=[0.3, 0.4, 0.3])

# Variable 2: Acceso a internet
opciones_internet = ["Sí", "No"]
acceso_internet = np.random.choice(opciones_internet, size=n, p=[0.7, 0.3])

# Crear DataFrame
data = pd.DataFrame({
    "Nivel Educativo": nivel_educativo,
    "Acceso a Internet": acceso_internet
})


# Tabla de contingencia

tabla = pd.crosstab(data["Nivel Educativo"], data["Acceso a Internet"])


# Prueba Chi-cuadrado

chi2, p, dof, expected = chi2_contingency(tabla)


# Visualización con mapa de calor

plt.figure(figsize=(8,6))
sns.heatmap(tabla, annot=True, fmt="d", cmap="Blues")
plt.title("Distribución conjunta: Nivel educativo vs. Acceso a Internet")
plt.ylabel("Nivel educativo")
plt.xlabel("Acceso a Internet")
plt.show()


# Resultados

print("Tabla de contingencia:\n", tabla, "\n")
print(f"Chi-cuadrado: {chi2:.2f}")
print(f"Grados de libertad: {dof}")
print(f"P-valor: {p:.4f}")
print("Frecuencias esperadas:\n", expected)
```

Distribución observada:

Nivel Básico: 4 sin internet, 14 con internet.

Nivel Medio: 3 sin internet, 18 con internet.

Nivel Superior: 5 sin internet, 16 con internet.

Se ve claramente que, en todos los niveles educativos, la mayoría tiene acceso a internet.

Interpretación del mapa de calor:

Los cuadros más oscuros están en la columna “Sí” (acceso a internet), lo que muestra que la mayor parte de los estudiantes/personas accede a internet.

El nivel Medio con internet (18) es el grupo más numeroso.

Los casos sin internet son pocos en comparación, independientemente del nivel educativo.

Prueba Chi-cuadrado (independencia):

Con esta tabla se calcula el estadístico χ² para verificar si el acceso a internet depende del nivel educativo.

Generalmente, si el p-valor > 0.05, se concluye que no hay relación significativa → el acceso a internet no está condicionado por el nivel educativo.

Si el p-valor ≤ 0.05, se concluye que sí existe asociación → el acceso a internet varía según el nivel educativo.

La mayoría de las personas, sin importar su nivel educativo, tiene acceso a internet.

Aunque el acceso parece un poco más alto en niveles Medio y Superior, la diferencia no es muy marcada.

El test Chi-cuadrado permitiría confirmar si esta diferencia es estadísticamente significativa o no.

En la práctica, esto refleja que hoy en día el internet es bastante común en todos los niveles educativos, aunque aún existen pequeños grupos sin acceso.
