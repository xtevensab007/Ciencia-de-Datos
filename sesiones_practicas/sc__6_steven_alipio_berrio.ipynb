iMPORTACION DE DATOS:

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sofía'],
    'Edad': [25, np.nan, 35, 29, -5, 25, None],
    'Ciudad': ['Bogotá', 'Medellín', None, 'Medellín', 'Cali', 'Bogotá', 'Cali'],
    'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
    'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10', '2023-01-12', None, '2023-01-15']
})
```

informacion de las variables presentes en el dataset

```python
df.info()
```

Descripcion de las variables

```python
df.describe()
```

identificacion de variables faltantes

```python
df.isnull().sum()
```

Imputacion de variable edad para corregir valores faltantes o icorrectos utilizando el valor de la media como referencia

```python
df['Edad'] = df['Edad'].apply(lambda x: np.nan if x < 0 else x)
```

Imputar edad  (media)

```python
edad_media = df['Edad'].mean()
df['Edad'] = df['Edad'].fillna(edad_media)
```

imputar ingreso

imputacion de la variable ingresos para corregir valores faltantes o icorrectos utilizando el valor de la mediana como referencia para el cambio

```python
ingreso_mediana = df['Ingreso'].median()
df['Ingreso'] = df['Ingreso'].fillna(ingreso_mediana)
```

imputar ciudad

Imputacion de la variable ciudad para corregir valores faltantes utilizando la moda como referencia para la correccion

```python
ciudad_moda = df['Ciudad'].mode()[0]
df['Ciudad'] = df['Ciudad'].fillna(ciudad_moda)
```

nombre y fecha de ingreso

```python
df['Nombre'] = df['Nombre'].fillna('Desconocido')
fecha_moda = df['FechaIngreso'].mode()[0]
df['FechaIngreso'] = df['FechaIngreso'].fillna(fecha_moda)
```

Resultados despues de imputaciones

```python
df.isnull().sum()
```

```python
import pandas as pd

# Crear el DataFrame base
df = pd.DataFrame({
    'ID': [101, 102, 102, 103, 104, 104, 104],
    'Nombre': ['Ana', 'Luis', 'Luis', 'Marta', 'Carlos', 'Carlos', 'Carlos'],
    'Edad': [25, 30, 30, 29, 40, 40, 41],
    'Ciudad': ['Bogotá', 'Cali', 'Cali', 'Medellín', 'Cali', 'Cali', 'Cali'],
    'FechaRegistro': ['2023-01-01', '2023-01-05', '2023-01-05', '2023-01-10',
                      '2023-01-15', '2023-01-15', '2023-01-16']
})

# Mostrar el DataFrame
df
```

1:¿Cuál es el total de registros originales?

```python
total_registros = len(df)
print("Total de registros originales:", total_registros)
```

2: ¿Cuáles y cuántos son los duplicados exactos?

```python
# Detectar duplicados exactos (todas las columnas)
duplicados_exactos = df[df.duplicated(keep=False)]
cantidad_duplicados_exactos =duplicados_exactos.shape[0]

print("Duplicados exactos encontrados:",duplicados_exactos)

print("Cantidad de duplicados exactos:", cantidad_duplicados_exactos)
```

3: ¿Cuáles y cuántos son los duplicados por varias columnas?

```python
# Detectar duplicados por columnas clave
duplicados_parciales = df[df.duplicated(subset=['ID', 'Nombre'], keep=False)]
cantidad_duplicados_parciales = duplicados_parciales.shape[0]

print("Duplicados por ['ID', 'Nombre']:")
print(duplicados_parciales)
print("Cantidad de duplicados por columnas:", cantidad_duplicados_parciales)
```

4: ¿Cuántos registros debes eliminar?

```python
# Si eliminamos duplicados por 'ID' y 'Nombre', dejando solo el primero
registros_a_eliminar = df.duplicated(subset=['ID', 'Nombre']).sum()
print("Registros a eliminar:", registros_a_eliminar)
```

5: ¿Cuántos registros quedan después de la limpieza?

```python
# Eliminar duplicados por 'ID' y 'Nombre'
df_limpio = df.drop_duplicates(subset=['ID', 'Nombre'])

# Mostrar total restante
print("Registros después de la limpieza:", len(df_limpio))

```

```python
df = pd.DataFrame({
    'Ciudad': ['bogota', 'Bogotá', 'BOGOTA', 'bogotá', 'bogata', 'Bógota', 'BogoTa', 'Cali', 'calí', 'medellín', 'medellin']
})
```

```python
from unidecode import unidecode

# Crear nueva columna normalizada (sin tildes, minúsculas)
df['Ciudad_normalizada'] = df['Ciudad'].apply(lambda x: unidecode(x).lower())

# Mostrar resultados
df
```

```python
# Diccionario de mapeo para estandarizar
mapeo_ciudades = {
    'bogota': 'Bogotá',
    'bogata': 'Bogotá',
    'cali': 'Cali',
    'calí': 'Cali',
    'medellin': 'Medellín'
}

# Aplicar mapeo
df['Ciudad_estandar'] = df['Ciudad_normalizada'].map(mapeo_ciudades)

# Mostrar resultados
df
```
